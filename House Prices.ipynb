```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
train=pd.read_csv(r"F:\Menna\Datasets\house-prices-advanced-regression-techniques\train.csv")
test=pd.read_csv(r"F:\Menna\Datasets\house-prices-advanced-regression-techniques\test.csv")
```

```python
train.isnull().sum()
```

```python
train=train.drop('Id',axis=1)
test=test.drop('Id',axis=1)
```

```python
nums=['int64','float64']
train=train.select_dtypes(include=nums)
test=test.select_dtypes(include=nums)
```

```python
train.isnull().sum()
```

```python
train=train.loc[:,train.isnull().sum()==0]
test=test.loc[:,train.columns.drop(['SalePrice'])]
```

```python
from sklearn.model_selection import train_test_split
X=train[train.columns.drop(['SalePrice'])]
Y=train[['SalePrice']]
X_train,X_val,Y_train,Y_val=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
print("original train shape:",X.shape)
print()
print("X_train shape",X_train.shape)
print("X_val shape",X_val.shape)
print()
print("Y_train shape",Y_train.shape)
print("Y_val shape",Y_val.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_sc=sc.fit_transform(X_train)
X_val_sc=sc.transform(X_val)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train_sc,Y_train)
```

Evaluation

```python
Y_pred=model.predict(X_val_sc)
```

```python
Y_pred
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(Y_val,Y_pred)
print("mean square error = ",MSE)
print("root mean squared error = ",np.sqrt(MSE))
```
